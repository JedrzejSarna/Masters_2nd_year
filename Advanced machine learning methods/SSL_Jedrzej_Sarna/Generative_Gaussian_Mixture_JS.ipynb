# Generative Gaussian Mixture

Możemy teraz zaimplementować model w Pythonie przy użyciu prostego dwuwymiarowego zbioru danych, utworzonego za pomocą funkcji make_blobs() dostarczonej przez scikit-learn. 
Funkcja ta umożliwia tworzenie syntetycznych zbiorów danych do testowania algorytmów, które wymagają punktów danych pobranych z zestawu rozkładów normalnych. 

Celem tego przykładu jest pokazanie dynamiki Generative Gaussian Mixture, dlatego dobrowolnie uniknęliśmy bardziej złożonych zbiorów danych, których nie można łatwo wizualizować.

```python
from sklearn.datasets import make_blobs #generuje zbior danych
import numpy as np

np.random.seed(1000)

nb_samples = 250 # liczba oznaczonych danych
nb_unlabeled = 200 # liczba nieoznaczonych danych

X, Y = make_blobs(n_samples=nb_samples, n_features=2, centers=2
                  ,cluster_std=1.25, random_state=100)
#n_samples -> liczba puntkow do wygenerowania
#n_features -> liczba cech do kazdego punktu
#centers -> liczba klastrow (klas)
#cluster_std -> odchylenie standardowe klastra
#random_state -> inicjowanie generatora liczb losowych, ustawiony na 100, aby uzyskac te same dane

#X -> macierz cech
#Y -> etykiety przypisane danym (mamy dwie klasy)

unlabeled_idx = np.random.choice(np.arange(0, nb_samples, 1)
                                 ,replace=False, size=nb_unlabeled) #losowy wybor indeksow probek bez etykiet
#np.arange -> tworzy wektor indeksow
#replace -> wartosc Boolean czy indeksy sa zwracane
#size -> wielkosc zbioru
```

Utworzyliśmy 250 próbek należących do 2 klas. Następnie losowo wybrano 200 punktów, które staną się naszym nieoznakowanym zbiorem danych (odpowiednia klasa została ustawiona na -1). Możemy teraz zainicjować dwa rozkłady gaussowskie, definiując ich średnią, kowariancję i wagę. Jedną z możliwości jest użycie losowych wartości, jednak my ustalimy już wartości aby uniknąć potencjalnych błędów

W przypadku modelu Gaussian Mixture Model i wielu innych algorytmów statystycznych, macierze kowariancji muszą spełniać pewne założenia.

Symetria: Macierze kowariancji muszą być symetryczne. Symetryczność jest wymagana, ponieważ kowariancja między zmiennymi $X$ a $Y$ jest taka sama jak między $Y$ a $X$. W macierzy kowariancji, elementy leżące po obu stronach głównej przekątnej muszą być równe.

Dodatnia określoność: Macierze kowariancji muszą być dodatnio określone. Oznacza to, że dla dowolnego niezerowego wektora $v$, iloczyn $v^T \Sigma v$ (gdzie $\Sigma$ to macierz kowariancji) musi być dodatni. To jest ważne, ponieważ gwarantuje, że macierz kowariancji jest odwracalna.



Nie należy się martwić sztywno ustawionymi parametrami, gdyż algorytm przesunie je i zdeformuje, aby dopasować je do rozkładu wejściowego.

```python
m1 = np.array([-2.0, -2.5])
c1 = np.array([[1.0, 1.0],
              [1.0, 2.0]])
q1 = 0.5


m2 = np.array([1.0, 3.0])
c2 = np.array([[2.0, -1.0],
              [-1.0, 3.5]])
q2 = 0.5
```

Jeśli rzutujemy gaussiany na płaszczyznę xy i obcinamy je, ograniczając zakres zmiennych niezależnych, można zobaczyć, że pojawiają się one jako elipsy. Aby określić ich strukturę i orientację, musimy zaobserwować, że:

 - Oś wielka jest określona jako wektor własny związany z największą wartością własną macierzy kowariancji. W ten sam sposób oś mała jest określona jako wektor własny związany z najmniejszą wartością własną (w przypadku wielowymiarowym konieczne jest przyjęcie wartości własnych w porządku malejącym).

 - Mimośród e jest równy stosunkowi dwóch wartości własnych. Gdy e = 1, elipsa jest okręgiem, ponieważ obie osie mają równą długość. Gdy e różne od 1, elipsa jest elipsą właściwą, rozciągniętą wzdłuż jednej z dwóch osi.
 
 Wyliczmy zatem  wartości oraz wektory własne dla naszych macierzy kowariancji.

```python
w1, v1 = np.linalg.eigh(c1) #wartosci wlasne i wektory wlasne macierzy kowariancji
print(w1, v1)
print('\n')
w2, v2 = np.linalg.eigh(c2)
print(w2, v2)


nv1 = v1 / np.linalg.norm(v1) #normalizacja wektora
nv2 = v2 / np.linalg.norm(v2)
```

Pozostało nam obliczyć kąt orientacji elipsy $\alpha$. Jest to kąt, pod jakim jest obrócona główna oś elipsy względem osi referencyjnej (najczęściej osi poziomej lub pionowej, u nas będzie to oś pozioma). Korzystamy ze znanego wzoru, wykorzystującego iloczyn skalarny:
$\alpha = \arccos\left(\frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}\right)$


```python
a1 = np.arccos(np.dot(nv1[:, 1], [1.0, 0.0]) / np.linalg.norm(nv1[:,1])) * 180.0 / np.pi
#obliczany kat miedzy wektorem wlasnym a osia x (okresla nachylenie elipsy na wykresie)

a2 = np.arccos(np.dot(nv2[:, 1], [1.0, 0.0]) / np.linalg.norm(nv2[:,1])) * 180.0 / np.pi

print(a1,a2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.core.display import HTML
display(HTML("<style>div.output_scroll { height: 60em; }</style>"))

from matplotlib.patches import Ellipse
sns.set()


def show_dataset():
    w1, v1 = np.linalg.eigh(c1)
    w2, v2 = np.linalg.eigh(c2)

    nv1 = v1 / np.linalg.norm(v1)
    nv2 = v2 / np.linalg.norm(v2)

    a1 = np.arccos(np.dot(nv1[:, 1], [1.0, 0.0]) / np.linalg.norm(nv1[:, 1])) * 180.0 / np.pi
    a2 = np.arccos(np.dot(nv2[:, 1], [1.0, 0.0]) / np.linalg.norm(nv2[:, 1])) * 180.0 / np.pi

    fig, ax = plt.subplots(figsize=(20, 15))

    ax.scatter(X[Y == 0, 0], X[Y == 0, 1], s=120, marker='o', label='Class 1') 
    #tworzy wykres punktowy punktow z klasy 1
    ax.scatter(X[Y == 1, 0], X[Y == 1, 1], s=120, marker='d', label='Class 2')
    ax.scatter(X[Y == -1, 0], X[Y == -1, 1], s=120, marker='x', label='Unlabeled')

    g1 = Ellipse(xy=m1, width=w1[1] * 3, height=w1[0] * 3, fill=False, linestyle='dashed', angle=a1, color='black',
                 linewidth=1) #tworzony obiekt elipsy, ktory szacuje gestosc dla kazdej z klas, centrowane wokol m1
    g1_1 = Ellipse(xy=m1, width=w1[1] * 2, height=w1[0] * 2, fill=False, linestyle='dashed', angle=a1, color='black',
                   linewidth=2)
    g1_2 = Ellipse(xy=m1, width=w1[1] * 1.4, height=w1[0] * 1.4, fill=False, linestyle='dashed', angle=a1,
                   color='black', linewidth=3)

    g2 = Ellipse(xy=m2, width=w2[1] * 3, height=w2[0] * 3, fill=False, linestyle='dashed', angle=a2, color='black',
                 linewidth=1)
    g2_1 = Ellipse(xy=m2, width=w2[1] * 2, height=w2[0] * 2, fill=False, linestyle='dashed', angle=a2, color='black',
                   linewidth=2)
    g2_2 = Ellipse(xy=m2, width=w2[1] * 1.4, height=w2[0] * 1.4, fill=False, linestyle='dashed', angle=a2,
                   color='black', linewidth=3)

    ax.set_xlabel(r'$x_0$', fontsize=18)
    ax.set_ylabel(r'$x_1$', fontsize=18)

    ax.add_artist(g1) #dodaje obiekty elips do wykresy
    ax.add_artist(g1_1)
    ax.add_artist(g1_2)
    ax.add_artist(g2)
    ax.add_artist(g2_1)
    ax.add_artist(g2_2)

    ax.legend(fontsize=18)

    plt.show()
```

```python
from scipy.stats import multivariate_normal

threshold = 1e-4

def total_norm():
    '''
    Oblicza sume odleglosci miedzy wektorami. 
    '''
    global m1, m1_old, m2, m2_old, c1, c1_old, c2, c2_old, q1, q1_old,q2, q2_old
    
    return np.linalg.norm(m1 - m1_old) + \
    np.linalg.norm(m2 - m2_old) + \
    np.linalg.norm(c1 - c1_old) + \
    np.linalg.norm(c2 - c2_old) + \
    np.linalg.norm(q1 - q1_old) + \
    np.linalg.norm(q2 - q2_old)
```

```python
#Inicjujemy parametry
m1_old = np.zeros((2,))
c1_old = np.zeros((2, 2))
q1_old = 0

m2_old = np.zeros((2,))
c2_old = np.zeros((2, 2))
q2_old = 0

# Proces treningowy
while total_norm() > threshold: #wykonuje sie dopoki roznica miedzy parametrami modelu przekracza ustalony prog
    
    #aktualizacja zmiennych na podstawie poprzednich wartosci
    m1_old = m1.copy()
    c1_old = c1.copy()
    q1_old = q1

    m2_old = m2.copy()
    c2_old = c2.copy()
    q2_old = q2

    Pij = np.zeros((nb_samples, 2)) # macierz zawierajaca prawdopodobienstwa przynaleznosci do klastrow

    # Krok E
    '''
    Krok E jest odpowiedzialny za oszacowanie prawdopodobieństw dla każdej próbki danych w odniesieniu do
    różnych klastrów, uwzględniając zarówno dane oznakowane,
    jak i nieoznakowane.
    '''
    for i in range(nb_samples): #iteruje przez wszystkie probki danych
        
        if Y[i] == -1: #jezeli probka nieoznakowana

            p1 = multivariate_normal.pdf(X[i], m1, c1, allow_singular=True) * q1
            #oblicza prawdpodobienstwo przynaleznosci do klastrow 0 i 1
            #allow_singular=True -> pozwala aby macierz kowariancji mogła być osobliwa (nieodwracalna)
            
            p2 = multivariate_normal.pdf(X[i], m2, c2, allow_singular=True) * q2
            
            Pij[i] = [p1, p2] / (p1 + p2) #wynik znormalizowany i zapisany w macierzy Pij
            
        else: #jezeli jest oznakowane
            
            Pij[i, :] = [1.0, 0.0] if Y[i] == 0 else [0.0, 1.0]
            #ustawia prawdopodobienstwa na podstawie etykiety

    # Krok M
    '''
    Krok M polega na aktualizacji parametrów modelu, takich jak
    wartości oczekiwane i macierze kowariancji dla klastrów oraz
    prawdopodobieństwa przynależności do klastrów. Te aktualizacje
    są dokonywane na podstawie oszacowań otrzymanych w Kroku E
    i mają na celu znalezienie takich parametrów, które maksymalizują
    spodziewaną wartość funkcji log-likelihood w modelu.
    '''
    n = np.sum(Pij, axis=0) #oblicza sumy prawdopodobienstw dla kazdego klastra (sumy kolumn)
    
    m = np.sum(np.dot(Pij.T, X), axis=0) #ważona suma cech próbek, gdzie wagi są określane przez prawdopodobieństwa
    #przynależności do klastrów jest cala macierz wartosci oczekiwanych

    m1 = np.dot(Pij[:, 0], X) / n[0] #aktualizuje wartosc oczekiwana dla klastra 0 
    m2 = np.dot(Pij[:, 1], X) / n[1] 

    q1 = n[0] / float(nb_samples) #aktualizuje prawdopodobienstwo przynaleznosci do klastra 0
    q2 = n[1] / float(nb_samples)

    c1 = np.zeros((2, 2)) #iniciuje macierze kowariancji
    c2 = np.zeros((2, 2))

    for t in range(nb_samples):
        c1 += Pij[t, 0] * np.outer(X[t] - m1, X[t] - m1) #aktualizuje macierz kowariancji dla klastra 0
        
        c2 += Pij[t, 1] * np.outer(X[t] - m2, X[t] - m2)

    c1 /= n[0] #Dzieli macierz kowariancji klastra 0 przez liczbę próbek przypisanych do tego klastra
    c2 /= n[1]
```
